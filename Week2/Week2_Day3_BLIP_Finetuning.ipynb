```python
import pandas as pd
import random

image_folder = "flickr8k_data/Images"
caption_file = "flickr8k_data/captions.txt"

captions_df = pd.read_csv(caption_file)

# Har image ka pehla caption hi use karenge training ke liye (simplicity ke liye)
captions_df = captions_df.drop_duplicates(subset="image", keep="first").reset_index(drop=True)

random.seed(42)
all_images = captions_df["image"].tolist()

# Chunki 8GB Colab free tier pe poora dataset (8000+ images) fine-tune karna slow hoga,
# hum ek subset lenge: 1000 images training ke liye
train_images = random.sample(all_images, 1000)
train_df = captions_df[captions_df["image"].isin(train_images)].reset_index(drop=True)

print("Training samples:", len(train_df))
print(train_df.head())
```
