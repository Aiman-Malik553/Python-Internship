```python
captions_path = "data/captions.txt"

with open(captions_path, 'r', encoding='utf-8') as f:
    lines = f.readlines()

print(f"Total lines: {len(lines)}")
print("\nFirst 5 lines:")
for line in lines[:5]:
    print(line.strip())
```

```python
# Pipe-separated file ko pandas DataFrame mein load karo
df = pd.read_csv(captions_path, sep='|')

# Column names ke extra spaces clean karo (agar hon)
df.columns = df.columns.str.strip()

print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
# Total unique images
num_images = df['image_name'].nunique()
print(f"Total unique images: {num_images}")

# Total captions
num_captions = df.shape[0]
print(f"Total captions: {num_captions}")

# Captions per image (average)
captions_per_image = df.groupby('image_name').size()
print(f"\nCaptions per image - stats:")
print(captions_per_image.describe())
```

```python
# Har caption ki length (words mein) nikalo
df['caption_length'] = df['caption_text'].apply(lambda x: len(str(x).split()))

print("Caption length stats (in words):")
print(df['caption_length'].describe())

# Histogram plot karo
plt.figure(figsize=(8,5))
plt.hist(df['caption_length'], bins=20, color='skyblue', edgecolor='black')
plt.title("Caption Length Distribution")
plt.xlabel("Number of words in caption")
plt.ylabel("Frequency")
plt.show()
```

```python
# Saari captions ko chota (lowercase) kar ke words mein tor do
all_words = []
for caption in df['caption_text']:
    words = str(caption).lower().split()
    all_words.extend(words)

# Unique words (vocabulary)
vocab = set(all_words)

print(f"Total words (with repetition): {len(all_words)}")
print(f"Vocabulary size (unique words): {len(vocab)}")

# Top 10 sabse common words dekho
from collections import Counter
word_freq = Counter(all_words)
print("\nTop 10 most common words:")
for word, freq in word_freq.most_common(10):
    print(f"{word}: {freq}")
```

```python
import random

# 3 random images choose karo unke captions ke sath
sample_images = random.sample(list(df['image_name'].unique()), 3)

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for i, img_name in enumerate(sample_images):
    img_path = os.path.join("data/images", img_name)
    img = Image.open(img_path)
    
    # Us image ki saari captions nikalo
    captions = df[df['image_name'] == img_name]['caption_text'].tolist()
    caption_text = "\n".join(captions[:3])  # pehli 3 captions dikhao
    
    axes[i].imshow(img)
    axes[i].axis('off')
    axes[i].set_title(caption_text, fontsize=9, wrap=True)

plt.tight_layout()
plt.show()
```

```python
# 1. Duplicate captions check karo
duplicate_captions = df[df.duplicated(subset=['caption_text'], keep=False)]
print(f"Duplicate captions count: {duplicate_captions.shape[0]}")

# 2. Blank ya bohot chhoti captions check karo (jaise 1-2 words wali)
short_captions = df[df['caption_length'] <= 2]
print(f"\nShort/blank captions (<=2 words): {short_captions.shape[0]}")
print(short_captions[['image_name', 'caption_text']].head())

# 3. Corrupted images check karo
corrupted_images = []
for img_name in df['image_name'].unique():
    img_path = os.path.join("data/images", img_name)
    try:
        img = Image.open(img_path)
        img.verify()  # check karega ke image valid hai ya nahi
    except Exception as e:
        corrupted_images.append(img_name)

print(f"\nCorrupted images found: {len(corrupted_images)}")
if corrupted_images:
    print(corrupted_images)
```
